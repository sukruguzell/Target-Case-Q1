```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and sort the dataset
df = pd.read_csv('Re_ Evreka Case Study/navigation_records.csv', parse_dates=['recorded_at'])
df = df.sort_values(by=['route_id', 'recorded_at'])

# Get unique route IDs
route_ids = df['route_id'].unique()


for route_id in route_ids:
    plt.figure(figsize=(12, 6))

    # Filter data for each route
    route_data = df[df['route_id'] == route_id]
    
    # Plot the distance over time for the route
    plt.plot(route_data['recorded_at'], route_data['distance'], marker='o', label=f'Route {route_id}')

    plt.gca().xaxis.set_major_locator(plt.matplotlib.dates.MinuteLocator(interval=30))  # Set 30-min interval
    plt.gca().xaxis.set_major_formatter(plt.matplotlib.dates.DateFormatter('%H:%M'))  # Format as HH:MM
    # Add labels, title, and legend
    plt.xlabel('Time')
    plt.ylabel('Distance')
    plt.title('Distance Over Time for Each Route')
    plt.legend(title='Route ID')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
    plt.grid(True)

    # Show the plot
    plt.tight_layout()
    plt.show()


```
